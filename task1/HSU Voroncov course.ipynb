```python
# titanic definitions
# run
import pandas
dataframe = pandas.read_csv('titanic.csv', index_col="PassengerId")

def writeFile(fName, data):
    with open(fName, 'w') as f:
        print(data, file=f, end='')
        
dataframe.head(13)
```

```python
# task 1: gender analysis  
import numpy as np

def countGender(gender):
    return len(dataframe[(dataframe.Sex == gender)]) 

writeFile('11.txt', str(countGender('male')) + ' ' + str(countGender('female') ))
```

```python
# task 2: percentage of survived
dfLen = len(dataframe.index)
survivedLen = len( dataframe[dataframe.Survived == 1] )
writeFile('2.txt', round((survivedLen / dfLen ) * 100 , 2) )
```

```python
# task 3: percentage of First Class passengers 
dfLen = len(dataframe.index)
firstClassLen = len(dataframe[dataframe.Pclass == 1])
writeFile('3.txt', round( ( firstClassLen / dfLen ) * 100, 2 ) )
```

```python
# https://towardsdatascience.com/filtering-data-frames-in-pandas-b570b1f834b9
type(dataframe.Age)
HEAD = 5
def headDframe(df):
    return df.head(HEAD)
headDframe( dataframe[['Age', 'Name']] )
headDframe( dataframe.select_dtypes(include=np.number) )
headDframe( dataframe.select_dtypes(include='number') )
dataframe[['Age', 'Name']].dtypes
dataframe[['Age', 'Name']].info()
dataframe.index
dataframe.loc[2] # selecting series.Series
dataframe.loc[[2]] # selecting single row frame.DataFrame
dataframe.iloc[1, [2,3,4]]
dataframe.loc[3, ['Age', 'Ticket']]
dataframe.loc[2:5, ['Age', 'Ticket']]
```

```python
# task 4: percentage of First Class passengers 
dfAge = dataframe[['Age']] # DataFrame of one column
# dfAge = dfAge.dropna() # cleaned from NaNs
 
# mean
mean = dfAge.mean(skipna = True, numeric_only= True) # DataFrame.mean() = sum / count
mean = mean.iloc[0] # to series.Series convertion
mean = round( mean, 2 ) 

# median
median = dfAge.median(skipna = True, numeric_only= True) # DataFrame.median() = devides a sorted array in halfs  
median = median.iloc[0] # to series.Series convertion
median = round( median, 2 ) 

# log
writeFile('4.txt', str( mean ) + ' ' + str( median ) )
```

```python
# task 5: pearson correlation
# alretnative way: 
# dataframe: dataframe[['A','B']]

pearson = dataframe['SibSp'].corr(dataframe['Parch'])
writeFile('5.txt', round( pearson, 2 ))
```
